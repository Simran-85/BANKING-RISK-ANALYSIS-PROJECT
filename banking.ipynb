```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
bins = [0, 100000,300000,  float('inf')]
labels = ['Low', 'Med', 'High']

df["Income Band"] = pd.cut(df["Estimated Income"], bins=bins, labels=labels, right=False)

df['Income Band'].value_counts().plot(kind="bar")
```

```python
# Examine the distribution of unique categories in categorical columns
categorical_cols = df[["BRId","GenderId","Risk Weighting","Nationality","Amount of Credit Cards","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Occupation","Income Band"]].columns
for col in categorical_cols:

  print(f"\nValue Counts for '{col}':")
  display(df[col].value_counts().plot(kind="bar"))
```

```python
#Univarite analysis
for i, predictor in enumerate(df[["BRId","GenderId","Risk Weighting","Nationality","Amount of Credit Cards","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Occupation","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
#Bivariate analysis
for i, predictor in enumerate(df[["BRId","GenderId","Risk Weighting","Nationality","Amount of Credit Cards","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Occupation","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
for i, predictor in enumerate(df[["BRId","GenderId","Risk Weighting","Nationality","Amount of Credit Cards","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Occupation","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
#Histoplot of value counts for different occupation
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show() 
```

```python
#numerical analysis

numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

#univariate analaysis and visualization

plt.figure(figsize=(25,20))
for i,col in enumerate(numerical_cols):
   plt.subplot(4,3,i+1)
   sns.histplot(df[col],kde=True)
   plt.title(col)
plt.show()   
```

```python
#Heatmaps
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
           
```

```python
#insights of EDA 

1.the strongest positive correlation occur among "Bank Deposists" with "Checking Accounts","Saving Accounts" and "foreign Currency Account" 
indicating  that customers who maintain hgh balances in one account type often hold substainal amount/funda across other accounts as well.
```
